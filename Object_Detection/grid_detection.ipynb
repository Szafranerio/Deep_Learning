```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('../DATA/flat_chessboard.png')
plt.imshow(flat_chess)
```

```python
found, corners = cv2.findChessboardCorners(flat_chess, (7,7))
if found == True:
    print('true')
else:
    print('false')
```

```python
#Cornersc
flat_chess = cv2.drawChessboardCorners(flat_chess, (7,7), corners, found)
plt.imshow(flat_chess)
```
