```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
#Mediam value of pxl
med = np.median(img)
med
```

```python
#Lower threshold  either 0 or 70% median
lower = int(max(0, 0.7*(med)))
#Upper threshold either 30% of median or x of 255
upper = int(min(255, 1.3*(med)))
```

```python
edges = cv2.Canny(img, threshold1=lower, threshold2=upper+350)
plt.imshow(edges)
```

```python
#Bluring
blurred = cv2.blur(img, ksize=(5,5))
edges = cv2.Canny(blurred, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
tennis = cv2.imread('../DATA/tennis.jpg')
tennis = cv2.cvtColor(tennis, cv2.COLOR_BGR2RGB)
plt.imshow(tennis)
```

```python
blurred_tennis = cv2.blur(tennis, (7,7))
edges = cv2.Canny(blurred_tennis, lower, upper+90)
plt.imshow(edges)
```
