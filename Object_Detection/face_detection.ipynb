```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
nadia = cv2.imread('../DATA/Nadia_Murad.jpg', 0)
denis = cv2.imread('../DATA/Denis_Mukwege.jpg', 0)
solvay = cv2.imread('../DATA/solvay_conference.jpg', 0)

def display(img):
    plt.imshow(img, cmap='gray')
```

```python
display(solvay)
```

```python
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detection(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)
    return face_img        
```

```python
result_denis = face_detection(denis)
plt.imshow(result_denis, cmap='gray')
```

```python
result_nadia = face_detection(nadia)
plt.imshow(result_nadia, cmap='gray')
```

```python
def adj_face_detection(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)
    return face_img     
```

```python
result_solvay = adj_face_detection(solvay)
plt.imshow(result_solvay, cmap='gray')
```

```python
#Eye detection

eye_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml')

def eye_detection(img):
    face_img = img.copy()
    eye_rects = eye_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)
    return face_img      
```

```python
result_nadia = eye_detection(nadia)
plt.imshow(result_nadia, cmap='gray')
```

```python
result_denis = eye_detection(denis)
plt.imshow(result_denis, cmap='gray')
```

```python
result_bart = face_detection(bart)
plt.imshow(result_bart, cmap='gray')
```
