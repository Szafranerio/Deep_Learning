```python
import cv2
import numpy as np  
import matplotlib.pyplot as plt
```

```python
dark_horse = cv2.imread('../DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('../DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('../DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
#OpenCv in BGR format color
hist_values_horse = cv2.calcHist([dark_horse], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values_horse)
```

```python
hist_values_rainbow = cv2.calcHist([rainbow], channels=[2], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values_rainbow)
```

```python
hist_values_bricks = cv2.calcHist([blue_bricks], channels=[2], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values_bricks)
```

```python
img_horse = dark_horse

color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img_horse], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim(0,256)
    
plt.title('Histogram for Horse')
```

```python
img_rainbow = rainbow

color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img_rainbow], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim(0,256)
    
plt.title('Histogram for Rainbow')
```

```python
img_bricks = blue_bricks

color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img_bricks], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim(0,256)
    
plt.title('Histogram for Bricks')
```

```python
masked_img = cv2.bitwise_and(img,img,mask,mask)
show_masked_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask)
plt.imshow(show_masked_img)
```

```python
hist_values_mask_red = cv2.calcHist([rainbow], channels=[2], mask=mask, histSize = [256], ranges = [0,256])
plt.plot(hist_values_mask_red)
plt.title('Hist for masked rainbow')
```

```python
gorilla =cv2.imread('../DATA/gorilla.jpg', 0)
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
display_img(gorilla, cmap='gray')
```

```python
hist_val = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_val)
```

```python
eq_gor = cv2.equalizeHist(gorilla)
display_img(eq_gor, cmap='gray')
```

```python
hist_val_eq = cv2.calcHist([eq_gor], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_val_eq)
```
