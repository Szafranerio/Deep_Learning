```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text='Hello', org = (50,300),fontFace = font, fontScale = 5, color=(255,255,255), thickness = 25)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
#Errosion
kernel = np.ones((5,5), dtype=np.uint8)
err = cv2.erode(img, kernel, iterations = 4)
display_img(err)
```

```python
#Opening, good for reducing noise in the pictures
img = load_img()
white_noise = np.random.randint(low = 0, high = 2, size=(600,600))
white_noise = white_noise * 255
display_img(white_noise)
```

```python
noise_img = white_noise + img
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
display_img(opening)
```

```python
#Morp gradient (edge detection)
img = load_img()
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
